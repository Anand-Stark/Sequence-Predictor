```python
import pandas as pd
import numpy as np
import collections
from PIL import Image
import pickle
```

```python
df = pd.read_excel("RawData.xlsx")
df.head()
```

```python
df = df[df['SourceOrganism'] != 'unclassified']
```

```python
sequences = np.array(df['Sequence'])
classes = np.array(df['SourceOrganism'])
```

```python
print(sequences[0])
print(type(sequences[0]))
print(len(sequences[0]))
```

```python
maxi = 0
for i in range(len(sequences)):
    if len(sequences[i]) > maxi:
        maxi = len(sequences[i])
        
print(maxi)
```

```python
k=6

kmer_counter = collections.Counter()
for i in range(len(sequences)):
    for j in range(len(sequences[i])):
        kmer = sequences[i][j:j+k]
        kmer_counter[kmer] += 1
```

```python
print(len(kmer_counter))
```

```python
img_length = 64
most_common = kmer_counter.most_common(img_length * img_length)
print(most_common)
print(type(most_common))
print(len(most_common))
```

```python
most_common_dict = dict(most_common)
```

```python
# retrieving only most common of all the sequences
kmers_list = []
for i in range(len(sequences)):
    current_counter = collections.Counter()
    for j in range(len(sequences[i])):
        kmer = sequences[i][j:j+k]
        if kmer in most_common_dict:
            current_counter[kmer] += 1
        
    kmers_list.append(current_counter)
```

```python
print(kmers_list[0])
print(kmers_list[1])

print(len(kmers_list[1]))

print(len(kmers_list))
```

```python
print(dict(kmers_list[0]))
```

```python
kmers_list_new = list()
for i in range(len(kmers_list)):
    kmslist_i_dict = dict(kmers_list[i])
    for key in most_common_dict.keys():
        if key not in kmslist_i_dict.keys():
            kmslist_i_dict[key] = 0
            
    kmers_list_new.append(kmslist_i_dict)
```

```python
print(kmers_list_new[0])
print(kmers_list_new[1])
print(len(kmers_list_new))
print(len(kmers_list_new[0]))

for i in range(len(kmers_list_new)):
    if len(kmers_list_new[i]) != img_length*img_length:
        print('found not img_length!')
```

```python
def convert_sequence_into_kmer_representation(seq, k_value=6, img_dimension=64):
    kmer_counter = collections.Counter()
    for i in range(len(seq)):
        kmer = seq[i: i+k_value]
        kmer_counter[kmer] += 1
    
    kmer_counter_ordered = dict()
    for i in range(img_dimension*img_dimension):
        if most_common_kmers_list[i] in kmer_counter.keys():
            kmer_counter_ordered[most_common_kmers_list[i]] = kmer_counter[most_common_kmers_list[i]]
        else:
            kmer_counter_ordered[most_common_kmers_list[i]] = 0
            
    kmer_values = list(kmer_counter_ordered.values())
    kmer_values = np.array(kmer_values)
    kmer_values = kmer_values.reshape((64, 64))
    
    return kmer_values
```
