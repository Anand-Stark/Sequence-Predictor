```python
new_sequences = []
new_labels = []
counts = [0,0,0,0,0,0]

for i in range(len(sequences)):
    if counts[labels[i]] < 200:
        new_sequences.append(sequences[i])
        new_labels.append(labels[i])
        counts[labels[i]] += 1
        
sequences = np.array(new_sequences)
labels = np.array(new_labels)

print(np.unique(labels, return_counts=True))
print(sequences.shape, labels.shape)
```

```python
k=6

kmer_counter = collections.Counter()
for i in range(len(sequences)):
    for j in range(len(sequences[i])):
        kmer = sequences[i][j:j+k]
        kmer_counter[kmer] += 1
```

```python
print(len(kmer_counter))
```

```python
img_length = 32
most_common = kmer_counter.most_common(img_length * img_length)
print(most_common)
print(type(most_common))
print(len(most_common))
```

```python
most_common_dict = dict(most_common)
```

```python
# retrieving only most common of all the sequences
kmers_list = []
for i in range(len(sequences)):
    current_counter = collections.Counter()
    for j in range(len(sequences[i])):
        kmer = sequences[i][j:j+k]
        if kmer in most_common_dict:
            current_counter[kmer] += 1
        
    kmers_list.append(current_counter)
```

```python
def convert_sequence_into_kmer_representation(seq, k_value=6, img_dimension=32):
    kmer_counter = collections.Counter()
    for i in range(len(seq)):
        kmer = seq[i: i+k_value]
        kmer_counter[kmer] += 1
    
    kmer_counter_ordered = dict()
    for i in range(img_dimension*img_dimension):
        if most_common_kmers_list[i] in kmer_counter.keys():
            kmer_counter_ordered[most_common_kmers_list[i]] = kmer_counter[most_common_kmers_list[i]]
        else:
            kmer_counter_ordered[most_common_kmers_list[i]] = 0
            
    kmer_values = list(kmer_counter_ordered.values())
    kmer_values = np.array(kmer_values)
    kmer_values = kmer_values.reshape((img_length, img_length))
    
    return kmer_values
```

```python
def convert_to_image(img_array):
    img_array = ((img_array - img_array.min()) / (img_array.max() - img_array.min()) * 255).astype(np.uint8)
    
    img = Image.fromarray(img_array)
    img.save(f'./images_coronavirus_6class/image_{i}_{labels[i]}.png')

# labels = np.array(list(df['SourceOrganism']))
for i in range(len(sequences)):
    img_array = convert_sequence_into_kmer_representation(sequences[i])
    
    convert_to_image(img_array)
```
